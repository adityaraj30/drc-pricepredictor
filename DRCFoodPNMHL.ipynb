```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
usr = pd.read_csv('/content/drive/MyDrive/Datasets/Data_High_Low.csv')
```

```python
usr
```

```python
usr = usr[['Product', 'Year', 'Month','Open', 'High', 'Low', 'Close','NextVal', 'NextHigh', 'NextLow']]
```

```python
usr.Product.unique()
```

```python
usr = pd.get_dummies(usr, columns = ['Product'])  #One Hot Encoding
usr
```

```python
predictors = ["Year", "Month", 'Product_beans', 'Product_cassava', 'Product_cassava_flour', 'Product_fish', 'Product_fish_salted',
       'Product_fish_smoked', 'Product_maize', 'Product_maize_meal', 'Product_meat_beef', 'Product_meat_chicken',
       'Product_meat_goat', 'Product_oil', 'Product_plantains', 'Product_rice', 'Product_salt', 'Product_sugar',
       'Product_wheat_flour',"Open","High","Low"]
target = ["NextVal", "NextHigh", "NextLow"]
```

```python
train = usr[usr["Year"]<2020].copy()
test = usr[usr["Year"]>=2020].copy()
```

```python
X_train = train[predictors]
X_test = test[predictors]
y_train = train[target]
y_test = test[target]
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
w = reg.coef_[0]
b = reg.intercept_
```

```python
w
```

```python
b
```

```python
predictions = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(predictions, y_test)
```

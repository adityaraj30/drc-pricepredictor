```python
predictors = ["Year", "Month", 'Product_beans', 'Product_cassava', 'Product_cassava_flour', 'Product_fish', 'Product_fish_salted',
       'Product_fish_smoked', 'Product_maize', 'Product_maize_meal', 'Product_meat_beef', 'Product_meat_chicken',
       'Product_meat_goat', 'Product_oil', 'Product_plantains', 'Product_rice', 'Product_salt', 'Product_sugar',
       'Product_wheat_flour',"Open","High","Low"]
target = ["Close"]
```

```python
#X_train, X_test, y_train, y_test = train_test_split(usr[predictors], usr[target], random_state = 0,  test_size=0.25, shuffle=False)
```

```python
train = usr[usr["Year"]<2020].copy()
test = usr[usr["Year"]>=2020].copy()
```

```python
X_train = train[predictors]
X_test = test[predictors]
y_train = train[target]
y_test = test[target]
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
predictions = reg.predict(X_test)
```

```python
X_test["predictions"] = predictions
```
